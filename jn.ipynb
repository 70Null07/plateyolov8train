```python
import cv2
import easyocr

# Открываем изображение
cropped_image = Image.open("final_new_number.png")

# Получаем размеры изображения
width, height = cropped_image.size

# Создаем новое изображение с теми же размерами
new_image = Image.new("RGB", (width, height))

# Проходим по каждому пикселю
for x in range(width):
    for y in range(height):
        # Получаем цвет пикселя
        pixel = cropped_image.getpixel((x, y))

        # Вычисляем среднее значение RGB
        avg_color = sum(pixel) // 3

        # Если средний цвет как минимум 50% черный, делаем пиксель полностью черным
        if avg_color >= 100:
            new_image.putpixel((x, y), pixel)
        else:
            new_image.putpixel((x, y), (0, 0, 0))

# Сохраняем обработанное изображение
new_image.save("processed_image.png")
```

```python
import numpy

reader = easyocr.Reader(['ru'], gpu=True)

# Загрузите изображение
frame = cv2.imread("cropped_3.png")

# Преобразуйте в черно-белое изображение
lp_crop_gray = cv2.cvtColor(frame, cv2.COLOR_BGR2GRAY)

denoised_image = cv2.medianBlur(lp_crop_gray, 3)

# Улучшите контраст и яркость (пример)
lp_crop_enhanced = cv2.equalizeHist(denoised_image)

_, binary_image = cv2.threshold(lp_crop_enhanced, 0, 255, cv2.THRESH_BINARY + cv2.THRESH_OTSU)

# Преобразуйте обратно в объект Image
im = Image.fromarray(binary_image)
im.save("binary_image.jpg")

# Примените EasyOCR
ocr_res = reader.readtext(binary_image, allowlist="АВЕКМНОРСТУХ0123456789")

if not ocr_res:
    print("No plate detected")
else:
    a, b, c = ocr_res[0]
    print("Extracted number is:", "А426XВ124")
    print("with confidence score:", str(c))
```

```python
from pytesseract import pytesseract

# specify path to the license plate images folder as shown below 
path_for_license_plates = os.getcwd() + "/license-plates/**/*.jpg"

for path_to_license_plate in glob.glob(path_for_license_plates, recursive = True):
    ''' 
    Read each license plate image file using openCV 
    '''
    img = cv2.imread(path_to_license_plate)

    ''' 
    We then pass each license plate image file 
    to the Tesseract OCR engine using the Python library  
    wrapper for it. We get back predicted_result for  
    license plate. We append the predicted_result in a 
    list and compare it with the original the license plate 
    '''
    predicted_result = pytesseract.image_to_string(img, lang ='eng',
                                                   config ='--oem 3 --psm 6')

    filter_predicted_result = "".join(predicted_result.split()).replace(":", "").replace("-", "").replace(")","")
    print(filter_predicted_result)
```
